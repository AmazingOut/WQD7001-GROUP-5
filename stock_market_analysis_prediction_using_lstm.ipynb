## Closing Price

The closing price is the last price at which the stock is traded during the regular trading day. A stocks closing price is the standard benchmark used by investors to track its performance over time.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("new_600718.csv")
dates = df["Date"]
close_prices = df["Close"]

dates_last_10 = dates[-10:]
close_prices_last_10 = close_prices[-10:]

plt.plot(dates, close_prices)

plt.plot(dates_last_10, close_prices_last_10, color="orange")

plt.xticks(dates, rotation=90)

plt.title("Close Prices")
plt.xlabel("Date")
plt.ylabel("Close")

plt.show()
```
